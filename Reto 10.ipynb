### 1. Desarrolle un programa que permita realizar la suma/resta de matrices. El programa debe validar las condiciones necesarias para ejecutar la operación.

```python
def ingresar_matriz(filas, columnas):
    #Creamos una lista vación que representará la matriz.
    matriz = []
    for i in range(filas):
        #Creamos la matriz de tamaño mxn.
        #A las posiciones se les suma 1 para que el usuario las vea desde (1,1) y no (0,0).
        fila = [float(input(f"Valor en ({i+1},{j+1}): ")) for j in range(columnas)]
        matriz.append(fila)
    return matriz

def operar_matrices(matriz1, matriz2, operacion):
    #Se realiza la operación de matrices respectiva.
    return [[(matriz1[i][j] + matriz2[i][j]) if operacion == '+' else (matriz1[i][j] - matriz2[i][j]) for j in range(len(matriz1[0]))] for i in range(len(matriz1))]

def imprimir_matriz(matriz):
    for fila in matriz:
        print("\t".join([str(elem) for elem in fila]))

def main():
    #Determinamos el tamaño de la matriz y la creamos.
    filas, columnas = [int(x) for x in input("Ingrese filas y columnas: ").split()]
    print("Matriz 1:")
    matriz1 = ingresar_matriz(filas, columnas)
    print("Matriz 2:")
    matriz2 = ingresar_matriz(filas, columnas)
    operacion = input("Operación (+/-): ")
    #Mostramos la matriz operada final.
    resultado = operar_matrices(matriz1, matriz2, operacion)
    print("Resultado:")
    imprimir_matriz(resultado)

if __name__ == "__main__":
    main()
```

### 2. Desarrolle un programa que permita realizar el producto de matrices. El programa debe validar las condiciones necesarias para ejecutar la operación.

```python
def ingresar_matriz(filas, columnas):
    """Solicita al usuario que ingrese una matriz de tamaño filas x columnas."""
    return [[float(input(f"Valor en ({i+1},{j+1}): ")) for j in range(columnas)] for i in range(filas)]

def multiplicar_matrices(matriz1, matriz2):
    """Multiplica dos matrices si sus dimensiones son compatibles."""
    if len(matriz1[0]) != len(matriz2):
        print("Error: No se pueden multiplicar estas matrices.")
        return None
    return [[sum(matriz1[i][k] * matriz2[k][j] for k in range(len(matriz2))) for j in range(len(matriz2[0]))] for i in range(len(matriz1))]

def imprimir_matriz(matriz):
    """Imprime una matriz de forma clara."""
    for fila in matriz:
        print("\t".join(map(str, fila)))

def main():
    """Función principal."""
    filas_a, columnas_a = map(int, input("Filas y columnas de la primera matriz: ").split())
    matriz1 = ingresar_matriz(filas_a, columnas_a)
    filas_b, columnas_b = map(int, input("Filas y columnas de la segunda matriz: ").split())
    matriz2 = ingresar_matriz(filas_b, columnas_b)
    resultado = multiplicar_matrices(matriz1, matriz2)
    if resultado:
        print("Resultado:")
        imprimir_matriz(resultado)

if __name__ == "__main__":
    main()
```

### 3. Desarrolle un programa que permita obtener la matriz transpuesta de una matriz ingresada. El programa debe validar las condiciones necesarias para ejecutar la operación.

```python
def ingresar_matriz(filas, columnas):
    #Solicita al usuario que ingrese una matriz de tamaño filas x columnas.
    for i in range(filas):
        fila = []
        for j in range(columnas):
            valor = float(input(f"Valor en ({i+1},{j+1}): "))
            fila.append(valor)
        matriz.append(fila)
    return matriz

def transponer_matriz(matriz):
    #Devuelve la matriz transpuesta.
    filas, columnas = len(matriz), len(matriz[0])
    transpuesta = []
    for i in range(columnas):
        nueva_fila = []
        for j in range(filas):
            nueva_fila.append(matriz[j][i])
        transpuesta.append(nueva_fila)
    return transpuesta

def imprimir_matriz(matriz):
    #Imprime una matriz de forma clara.
    for fila in matriz:
        print("\t".join(map(str, fila)))

def main():
    #Función principal.
    filas, columnas = map(int, input("Filas y columnas de la matriz: ").split())
    matriz = ingresar_matriz(filas, columnas)
    print("Matriz ingresada:")
    imprimir_matriz(matriz)
    resultado = transponer_matriz(matriz)
    print("Matriz transpuesta:")
    imprimir_matriz(resultado)

if __name__ == "__main__":
    main()
```

### 4. Desarrollar un programa que sume los elementos de una columna dada de una matriz.

```python
def ingresar_matriz(filas, columnas):
    #Solicita al usuario que ingrese una matriz de tamaño filas x columnas.
    matriz = []
    for i in range(filas):
        fila = []
        for j in range(columnas):
            valor = float(input(f"Valor en ({i+1},{j+1}): "))
            fila.append(valor)
        matriz.append(fila)
    return matriz


def sumar_columna(matriz, columna):
    #Suma los elementos de una columna dada.
    if columna < 0 or columna >= len(matriz[0]):
        print("Error: Índice de columna fuera de rango.")
        return None
    suma = 0
    for fila in matriz:
        suma += fila[columna]
    return suma

def imprimir_matriz(matriz):
    #Imprime una matriz de forma clara.
    for fila in matriz:
        print("\t".join(map(str, fila)))

def main():
    #Función principal.
    entrada = input("Filas y columnas de la matriz: ").split()
    filas = int(entrada[0])
    columnas = int(entrada[1])
    matriz = ingresar_matriz(filas, columnas)
    print("Matriz ingresada:")
    imprimir_matriz(matriz)
    columna = int(input("Ingrese el índice de la columna a sumar: "))
    suma = sumar_columna(matriz, columna)
    if suma is not None:
        print(f"La suma de la columna {columna} es: {suma}")

if __name__ == "__main__":
    main()
```

### 5. Desarrollar un programa que sume los elementos de una fila dada de una matriz.

```python
def ingresar_matriz(filas, columnas):
    """Solicita al usuario que ingrese una matriz de tamaño filas x columnas."""
    matriz = []
    for i in range(filas):
        fila = []
        for j in range(columnas):
            valor = float(input(f"Valor en ({i+1},{j+1}): "))
            fila.append(valor)
        matriz.append(fila)
    return matriz

def sumar_fila(matriz, fila):
    """Suma los elementos de una fila dada."""
    if fila < 0 or fila >= len(matriz):
        print("Error: Índice de fila fuera de rango.")
        return None
    return sum(matriz[fila])

def imprimir_matriz(matriz):
    """Imprime una matriz de forma clara."""
    for fila in matriz:
        print("\t".join(map(str, fila)))

def main():
    """Función principal."""
    entrada = input("Filas y columnas de la matriz: ").split()
    filas = int(entrada[0])
    columnas = int(entrada[1])
    matriz = ingresar_matriz(filas, columnas)
    print("Matriz ingresada:")
    imprimir_matriz(matriz)
    fila = int(input("Ingrese el índice de la fila a sumar: "))
    suma = sumar_fila(matriz, fila)
    if suma is not None:
        print(f"La suma de la fila {fila} es: {suma}")

if __name__ == "__main__":
    main()
```
